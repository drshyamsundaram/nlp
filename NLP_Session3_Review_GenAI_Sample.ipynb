```python
import nltk
from textblob import TextBlob
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Define the review text
review_text = """
What can I say!
SIMPLY SPECTACULAR.
PUBG - smooth +90 FPS
Smashing all the benchmarks by some margin (140%~~)
4 SPEAKERS are like home theatre.
Battery backup is excellent
Could have been better than 20 wts charger
"""

# Calculate the number of characters
num_characters = len(review_text)

# Calculate n-grams
from nltk.util import ngrams
n = 3  # You can change n to 1, 2, or 3 for different n-grams
tokens = nltk.word_tokenize(review_text)
n_grams = list(ngrams(tokens, n))

# Calculate sentiment analysis
analysis = TextBlob(review_text)
sentiment_subjectivity = analysis.sentiment.subjectivity
sentiment_polarity = analysis.sentiment.polarity

# Calculate sentence-wise sentiment
sentences = nltk.sent_tokenize(review_text)
sentence_sentiments = []

for sentence in sentences:
    sentence_blob = TextBlob(sentence)
    sentence_subjectivity = sentence_blob.sentiment.subjectivity
    sentence_polarity = sentence_blob.sentiment.polarity
    sentence_sentiments.append({
        "Sentence": sentence,
        "Subjectivity": sentence_subjectivity,
        "Polarity": sentence_polarity
    })

# Generate a word cloud
wordcloud = WordCloud(width=800, height=400, max_words=30, background_color='white').generate(review_text)

# Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
plt.show()

# Create the results JSON
results_json = {
    "Number of characters": num_characters,
    "Number of n-grams": len(n_grams),
    "Sentiment subjectivity": sentiment_subjectivity,
    "Sentiment polarity": sentiment_polarity,
    "Sentence-wise sentiment": sentence_sentiments,
    "Top 30 wordcloud": wordcloud.words_
}

# Print the results JSON
import json
print(json.dumps(results_json, indent=2))
```
